```python
import numpy as np

n = int(input("Enter number of products: "))

prices = np.array([
    int(input(f"Enter price of product {i+1}: "))
    for i in range(n)
])

print("Product Prices:", prices)

total = 0
maximum = prices[0]
minimum = prices[0]

for price in prices:
    total += price

    if price > maximum:
        maximum = price

    if price < minimum:
        minimum = price

print("Total Cart Value =", total)
print("Most Expensive Product =", maximum)
print("Cheapest Product =", minimum)
```

```python
# Q1.
# create a 5*5 matrix of random integer(1-90)
# find mean values of row and column wise
# replace all even numbers with -1
# sort the array row and column wise
# find the row and colmun wise sum saparetly
# count the number of negatives in each column
import numpy as np

arr = np.random.randint(1,91,(5,5))
print(arr)

print("\nRow-wise Mean")
print(np.mean(arr,axis=1))

print("\nColumn-wise Mean")
print(np.mean(arr,axis=0))

arr = np.where(arr%2==0,-1,arr)
print("\nreplace even number with -1:")
print(arr)

rowSorted = np.sort(arr,axis=1)
print("\nsorted arrray row wise:")
print(rowSorted)

rowSorted = np.sort(arr,axis=0)
print("\nsorted arrray column wise:")
print(rowSorted)

print("\nrow-wise sum:")
print(np.sum(arr,axis=1))

print("\ncolumn-wise sum:")
print(np.sum(arr,axis=0))

# number of negative in each column"
count_col = np.sum(np.where(arr<0,1,0),axis=0)
print("\nnumber of negative in each column:")
print(count_col)
```

```python
# Q2.
# Inventory Control System

# Task:

# Create a 5*5 array(stock levels)
# Create another 5*5 array(demand)
# Compute:
# Remaining stock = stock - demand
# Replace:
# Negative values->0 (out of stock)
# Find:
# Totol shortage
# Products with zero stock
# Sort each row bassed on remaining stock

stock = np.random.randint(1,100,(5,5))
print("\nStock matrix")
print(stock)

demand = np.random.randint(1,100,(5,5))
print("\nDemand Matrix")
print(demand)

remaining = stock - demand
print("\nRemaining")
print(remaining)

remaining = np.where(remaining<0,0,remaining)
print("\nRemaining")
print(remaining)

total_shortage = np.sum(np.where(remaining<0,remaining*-1,0),axis=0)
print("\nTotal Shortage")
print(total_shortage)

zero_stock = np.sum(np.where(remaining==0,1,0),axis=0)
print("\nProducts with zero stock")
print(zero_stock)

sorted_remaining = np.sort(remaining,axis=1)
print("\nSorted Remaining")
print(sorted_remaining)




```
