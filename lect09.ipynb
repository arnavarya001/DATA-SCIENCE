```python
# 1. Destructor

# A Destructor is a special member function that is automatically called when an object is destroyed. It is used to free memory and release resources.

# Characteristics
# Same name as the class preceded by ~.
# No return type.
# No parameters.
# Called automatically.
# Example (C++)
# #include <iostream>
# using namespace std;

# class Student {
# public:
#     Student() {
#         cout << "Constructor Called" << endl;
#     }

#     ~Student() {
#         cout << "Destructor Called" << endl;
#     }
# };

# int main() {
#     Student s;
#     return 0;
# }
# Output
# Constructor Called
# Destructor Called
# Uses
# Free dynamically allocated memory.
# Close files.
# Release network/database connections.
# 2. Virtual Polymorphism

# Runtime Polymorphism is achieved using virtual functions. It allows a derived class function to override a base class function.

# Virtual Function

# A function declared with the virtual keyword in the base class.

# Example
# #include <iostream>
# using namespace std;

# class Animal {
# public:
#     virtual void sound() {
#         cout << "Animal makes sound" << endl;
#     }
# };

# class Dog : public Animal {
# public:
#     void sound() override {
#         cout << "Dog barks" << endl;
#     }
# };

# int main() {
#     Animal *a;
#     Dog d;

#     a = &d;
#     a->sound();

#     return 0;
# }
# Output
# Dog barks
# Advantages
# Achieves runtime polymorphism.
# Improves flexibility and reusability.
# Supports dynamic binding.
# 3. Different Types of Inheritance

# Inheritance allows one class to acquire properties and methods of another class.

# A. Single Inheritance

# One derived class inherits from one base class.

# #include <iostream>
# using namespace std;

# class A {
# public:
#     void display() {
#         cout << "Class A" << endl;
#     }
# };

# class B : public A {
# };

# int main() {
#     B obj;
#     obj.display();
# }
# B. Multiple Inheritance

# One derived class inherits from multiple base classes.

# #include <iostream>
# using namespace std;

# class A {
# public:
#     void showA() {
#         cout << "Class A" << endl;
#     }
# };

# class B {
# public:
#     void showB() {
#         cout << "Class B" << endl;
#     }
# };

# class C : public A, public B {
# };

# int main() {
#     C obj;
#     obj.showA();
#     obj.showB();
# }
# C. Multilevel Inheritance

# A class inherits from another derived class.

# #include <iostream>
# using namespace std;

# class A {
# public:
#     void showA() {
#         cout << "Class A" << endl;
#     }
# };

# class B : public A {
# };

# class C : public B {
# };

# int main() {
#     C obj;
#     obj.showA();
# }
# D. Hierarchical Inheritance

# Multiple classes inherit from a single base class.

# #include <iostream>
# using namespace std;

# class A {
# public:
#     void display() {
#         cout << "Base Class" << endl;
#     }
# };

# class B : public A {
# };

# class C : public A {
# };

# int main() {
#     B obj1;
#     C obj2;

#     obj1.display();
#     obj2.display();
# }
# E. Hybrid Inheritance

# Combination of two or more inheritance types.

# #include <iostream>
# using namespace std;

# class A {
# public:
#     void show() {
#         cout << "Class A" << endl;
#     }
# };

# class B : public A {
# };

# class C : public A {
# };

# class D : public B, public C {
# };

# int main() {
#     D obj;
# }
# Summary Table
# Type	Description
# Single	One base → One derived
# Multiple	Multiple bases → One derived
# Multilevel	Base → Derived → Derived
# Hierarchical	One base → Multiple derived
# Hybrid	Combination of inheritance types

# These notes are suitable for a 5–10 mark college assignment/homework.
```
