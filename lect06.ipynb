```python
import pandas as pd

df = pd.read_csv('student_data.csv')

```

```python
df.groupby('placement_status')['exam_score'].agg('mean')
```

```python
df.groupby('placement_status')['study_hours'].agg('mean')
```

```python
df.groupby('study_hours')['exam_score'].agg('mean')
```

```python
def grades(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'


df['grade'] = df['exam_score'].apply(grades)
df
```

```python
import pandas as pd

student = pd.DataFrame({
    'Name': ['Arnav', 'Rahul', 'Amit', 'Yash'],
    'department id': [0, 3, 5, 7]
}, index=[1, 2, 3, 4])

department = pd.DataFrame({
    'Marks': [85, 90, 78, 32],
    'department': ['cse', 'ece', 'mech', 'civil']
}, index=[1, 2, 3, 4])

result = student.join(department, how='inner')
display(result)
```
