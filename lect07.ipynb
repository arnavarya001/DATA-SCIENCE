```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "id":[1,2,3,4],
    'name':['Raja','Rani','Rahul','Rhea'],
    'age':[25,30,35,40]
})

df2 = pd.DataFrame({
    "id":[3,4,5,6],
    'name':['Rahul','Rhea','Ravi','Reena'],
    'age':[35,40,45,50]
})


x = pd.merge(df1,df2,on = 'id',how = 'inner')   # Inner Join
# print(x)

x.rename(columns = {'name_x':'name_df1','name_y':'name_df2'},inplace = True)
print(x)
```
