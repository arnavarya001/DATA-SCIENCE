```python
n = int(input("Enter number of scores: "))
score = np.array([int(input("Enter score: "))
for i in range(n)])
print("Scores:", score)
total = 0
maximum = score[0]
minimum = score[0]

for i in score:
    total += i
    if i > maximum:
        maximum = i
    if i < minimum:
        minimum = i

print("Sum =", total)
print("Max =", maximum)
print("Min =", minimum)
```

```python
import numpy as np

days = ["Monday", "Tuesday", "Wednesday",
        "Thursday", "Friday", "Saturday", "Sunday"]

scores = []

for day in days:
    s = int(input(f"Enter score for {day}: "))
    scores.append(s)

scores = np.array(scores)

print("\nScores:")
for i in range(len(days)):
    print(days[i], ":", scores[i])

max_score = scores[0]
min_score = scores[0]
max_day = days[0]
min_day = days[0]

for i in range(len(scores)):
    if scores[i] > max_score:
        max_score = scores[i]
        max_day = days[i]

    if scores[i] < min_score:
        min_score = scores[i]
        min_day = days[i]

print("\nHighest Score:", max_score, "on", max_day)
print("Lowest Score:", min_score, "on", min_day)
```
