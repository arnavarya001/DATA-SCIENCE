```python
lst = [3,4,1,"hello",'c',[2,4],2,3]
print(lst)
#append - add list
lst.append([2,3,4])
print(lst)
#extend - add seprate element
lst.extend([33,55,66])
print(lst)
#insert(index,value)
lst.insert(2,'df')
print(lst)
#pop()- remove last element
lst.pop()
print(lst)
#remove(value)- remove any value from the list
lst.remove(55)
print(lst)
#index(value)
print(lst.index('hello'))
#count(value) - count the no. of values in the list
print(lst.count(3))
# sort()
```

```python
lst = [24,7,"arnav",'apple',44]
print(lst)
lst.append([2,3,2344])
print(lst)
lst.extend([5654,5545235,65346])
print(lst)
lst.insert(2,'plane')
print(lst)


```

```python
#sort()
lst = [2,1,7,3,6,2,0]
lst.sort()
print(lst)
#sum,min,max,sorted
print(sum(lst))
print(min(lst))
print(max(lst))
#sorted() - return new list
print(sorted(lst))
```
